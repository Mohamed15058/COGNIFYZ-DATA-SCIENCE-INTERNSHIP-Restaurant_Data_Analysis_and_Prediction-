## taks 3

```python
import plotly.express as px
# Create a scatter map
fig = px.scatter_mapbox(
    df,
    lat='Latitude',
    lon='Longitude',
    hover_name='Restaurant Name',
    hover_data=['City', 'Cuisines', 'Aggregate rating'],
    color_discrete_sequence=['red'],  # Set marker color to red
    zoom=3  # Adjust the initial zoom level
)

# Update the layout to use an OpenStreetMap background
fig.update_layout(mapbox_style="open-street-map")

# Show the interactive map
fig.show()
```

```python
sns.countplot(y = df['City'], order=df.City.value_counts().iloc[:10].index)
plt.title('Distribution of Restaurants Across Cities')
```

```python
# Calculate the correlation between latitude, longitude, and ratings
correlation_matrix = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()

# Create a heatmap to visualize the correlation with a different color map (e.g., 'viridis')
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title("Correlation Between Restaurant's location and Rating")
```
